```python
X = df[['R&D Spend','Administration']]
X
```

```python
y = df['Profit']
y
```

```python
sns.regplot(x = 'R&D Spend', y ='Profit', data = df)
plt.title("R&D Spend Vs Profit")
```

```python
##model building r &D and Admin. spend as independent variables. 
#Correlation analysis 
c = df[['R&D Spend','Administration','Profit']].corr()
sns.heatmap(c,annot=True,cmap='Blues')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()  ##### Create a variable of Model(object)
##### We fill fit/train the Model
model_train = model.fit(X,y)
print("The Model training is Completed")
```

```python
model_train.coef_ ##Coefficients m1,m2
```

```python
model_train.intercept_ ##constant c
```

```python
model_train.score(X,y) ##R^2
```

```python
X1 = int(input("Enter the R&D spend: "))
X2 = int(input("Enter the Addministration spend: "))
y = model_train.coef_[0]*X1 + model_train.coef_[1]*X2 + model_train.intercept_
```

```python
y
print("Profit is : ", y) ##94% accuracy. 
```
