```python
toyo=pd.concat([df.iloc[:,2:38]],axis=1)
toyo
```

```python
##toyo[('color')]=('red':0,'blue':1,'green':2)
toyo['Color'].replace('Red','0', inplace=True)
toyo['Color'].replace('Blue','1', inplace=True)
toyo['Color'].replace('Green','2', inplace=True)
toyo['Color'].replace('Silver','3', inplace=True)
toyo['Color'].replace('Black','4', inplace=True)
toyo['Color'].replace('Grey','5', inplace=True)
toyo['Fuel_Type'].replace('Petrol','0', inplace=True)
toyo['Fuel_Type'].replace('Diesel','1', inplace=True)
toyo['Fuel_Type'].replace('CNG','2', inplace=True)

toyo
```

```python
X = toyo[['Age_08_04' , 'KM', 'HP','cc','Doors','Gears','Quarterly_Tax','Weight','Central_Lock','Powered_Windows','Power_Steering','Radio','Mistlamps','Sport_Model','Backseat_Divider','Metallic_Rim','Radio_cassette','Tow_Bar']]
X
```
